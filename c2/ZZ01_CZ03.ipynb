### Zadanie 4: Konstrukcja szeregów rodzielczych - bardziej skomplikowane tablice [10 pkt]

#### a) Forma tabelaryczna
Przygotuj szeregi rozdzielcze VEI w formie tablicy `vei_data` zawierającej następujące elementy:

> Indeks: VEI </br>
> Wartości VEI posortowane od najmniejszej do największej.

> Kolumny: 2-poziomowe nagłówki, dla każdego z nagłówków 1 poziomu powtarza się cały zestaw nagłówków 2 poziomu

>> Nagłówki 1 poziomu: All, Confirmed Eruption, Uncertain Eruption </br>
>> Całość danych oraz grupy wyznaczone na podstawie kolumny Eruption category</br>

>> Nagłówki 2 poziomu: f, cf, rf, crf </br>
>> Zliczenia poszczególnych kategorii (ang. frequency, absolute frequency), zliczenia skumulowane poszczególnych kategorii (ang. cumulative frequency, cumulative absolute frequency),  rf - częstość zliczeń poszczególnych kategorii (ang. relative frequency), crf - skumulowana częstość zliczeń poszczególnych kategorii (ang. cumulative relative frequency)

Wyświetl przygotowane tablicę w taki sposób, aby liczby w kolumnach f i cf oraz indeks tablicy VEI wyświetlały się bez miejsc po przecinku, a liczby z kolumn rf i crf z dokładnością do 4 miejsca po przecinku.

```python
# YOUR CODE HERE
def vei_dist(data):
    f = data.groupby('VEI').size()
    cf = f.cumsum()
    rf = f / f.sum()
    crf = rf.cumsum()
    
    df = pd.DataFrame({'f': f,'cf': cf,'rf': rf,'crf': crf})
    
    return df
```

```python
idx = pd.Index(range(8), name='VEI')

cols = pd.MultiIndex.from_product([
    ['All eruptions', 'Confirmed Eruptions', 'Uncertain Eruptions'],
    ['f', 'cf', 'rf', 'crf']
])

vei_data = pd.DataFrame(index=idx, columns=cols)

all_data = vei_dist(eruptions_data)
vei_data['All eruptions'] = all_data

confirmed_data = vei_dist(eruptions_data[eruptions_data['Eruption Category'] == 'Confirmed'])
vei_data['Confirmed Eruptions'] = confirmed_data

uncertain_data = vei_dist(eruptions_data[eruptions_data['Eruption Category'] == 'Uncertain'])
vei_data['Uncertain Eruptions'] = uncertain_data

vei_data.style.format({
    ('All eruptions', 'f'): '{:.0f}',
    ('All eruptions', 'cf'): '{:.0f}',
    ('All eruptions', 'rf'): '{:.4f}',
    ('All eruptions', 'crf'): '{:.4f}',
    ('Confirmed Eruptions', 'f'): '{:.0f}',
    ('Confirmed Eruptions', 'cf'): '{:.0f}',
    ('Confirmed Eruptions', 'rf'): '{:.4f}',
    ('Confirmed Eruptions', 'crf'): '{:.4f}',
    ('Uncertain Eruptions', 'f'): '{:.0f}',
    ('Uncertain Eruptions', 'cf'): '{:.0f}',
    ('Uncertain Eruptions', 'rf'): '{:.4f}',
    ('Uncertain Eruptions', 'crf'): '{:.4f}'
})
```

### Zadanie 5: Rodzaje metod wykrywania erupcji [12 pkt]

Przygotuj tablice `evidence_data` zawiarającą informacje o tym ile erupcji o poszczególnych wartościach VEI zostało potwierdzonych przy użyciu poszczególnych metod weryfikacji.
Tablica powinna zawierać następujące elementy:

> Indeks: 2-poziomowe nagłówki </br>

>> Nagłówki 1 poziomu: Kategorie dowodów </br>

>> Nagłówki 2 poziomu: Metody, które dostarczyły dowodów </br>

> Kolumny: VEI </br>
> Wartości VEI posortowane od najmniejszej do największej oraz kolumna "All", w której znajdują się dane dotyczące wszystkich erupcji, bez podziału na VEI.

Wyświetl przygotowane tablicę w taki sposób, aby zawarte w tablicy liczby (w tym wartości VEI w nagłówkach kolumn) wyświetlały się bez miejsc po przecinku.

```python
# YOUR CODE HERE
# Create multi-level index for evidence types and methods
evidence_categories = eruptions_data['Evidence Type'].unique()
evidence_methods = eruptions_data['Evidence Method'].unique()

categories_methods = []
for category in evidence_categories:
    methods = eruptions_data[eruptions_data['Evidence Type'] == category]['Evidence Method'].unique()
    for method in methods:
        categories_methods.append((category, method))

multi_index = pd.MultiIndex.from_tuples(categories_methods, names=['Evidence Type', 'Evidence Method'])
columns = list(range(8)) + ['All']

evidence_data = pd.DataFrame(index=multi_index, columns=columns)
evidence_data = evidence_data.fillna(0)

for idx in evidence_data.index:
    category, method = idx
    mask = (eruptions_data['Evidence Type'] == category) & (eruptions_data['Evidence Method'] == method)
    evidence_data.loc[idx, 'All'] = mask.sum()

    for vei in range(8):
        vei_mask = mask & (eruptions_data['VEI'] == vei)
        evidence_data.loc[idx, vei] = vei_mask.sum()

evidence_data = evidence_data.astype(int)
evidence_data.columns.name = 'VEI'

evidence_data
```

Wyniki zestaw na pojedynczej figurze (siatka wykresów 2x1). Na górnym wykresie przedstaw dane rozróżnione tylko po kątem VEI, a na dolnym wykresie przedstaw dane rozróżnione zarówno po VEI jak i rodzaju dowodu ich wystąpienia (Evidence Type). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:

```python
fig, axes = plt.subplots(2, 1, figsize=(15, 12), dpi=100)
fig.set_facecolor('white')

# Upper plot - only VEI
sns.stripplot(x='VEI', y='Start Year', data=eruptions_data, 
              ax=axes[0], jitter=True, alpha=0.5, size=4)
axes[0].set_title('Distribution of Eruptions by VEI Over Time', fontsize=16)
axes[0].set_ylabel('Year', fontsize=12)
axes[0].set_xlabel('VEI', fontsize=12)
axes[0].set_xlim(-0.5, 7.5)
axes[0].set_xticklabels([f'{int(i)}' for i in range(8)])

# Lower plot - VEI and Evidence Type
sns.stripplot(x='VEI', y='Start Year', hue='Evidence Type', 
              data=eruptions_data, ax=axes[1], 
              jitter=True, alpha=0.4, size=2,
              palette='deep', dodge=True)
axes[1].set_title('Distribution of Eruptions by VEI and Evidence Type Over Time', fontsize=16)
axes[1].set_ylabel('Year', fontsize=12)
axes[1].set_xlabel('VEI', fontsize=12)
axes[1].set_xlim(-0.5, 7.5)
axes[1].set_xticklabels([f'{int(i)}' for i in range(8)])

handles, labels = axes[1].get_legend_handles_labels()
axes[1].legend(handles=handles, labels=labels, 
               title='Evidence Type', fontsize=10, title_fontsize=10,
               loc='upper center', bbox_to_anchor=(0.5, -0.15), 
               ncol=6, frameon=True)

fig.suptitle('Eruptions by VEI and Evidence Type Over Time', fontsize=20,  y=1.005)
plt.tight_layout()
plt.show()
```
